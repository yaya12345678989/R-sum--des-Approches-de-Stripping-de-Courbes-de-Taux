```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from scipy.optimize import brentq, newton

#Importation des data 
euribor3m=pd.read_csv('/Users/yaniskharbouch/Desktop/Python librairy/euribor3M.csv',sep=';')
euribor6m=pd.read_csv('/Users/yaniskharbouch/Desktop/Python librairy/euribor6M.csv',sep=';')
eonia=pd.read_csv('/Users/yaniskharbouch/Desktop/Python librairy/eonia.csv',sep=';')

# Date de référence
reference_date = datetime(2025, 2, 3)

# Fonctions de base pour les calculs de dates et de fractions d'année
def convert_maturity_to_days(maturity_str):
    """Convertit une chaîne de maturité (ex: '1M', '2Y') en nombre de jours"""
    if 'D' in maturity_str:
        return int(maturity_str.replace('D', ''))
    elif 'W' in maturity_str:
        return int(maturity_str.replace('W', '')) * 7
    elif 'M' in maturity_str:
        return int(maturity_str.replace('M', '')) * 30
    elif 'Y' in maturity_str:
        return int(maturity_str.replace('Y', '')) * 365
    else:
        return 0

def fractionAnnee(start_days, end_days, convention='ACT/365'):
    """
    Calcule la fraction d'année entre deux dates selon une convention donnée.
    
    Parameters:
    -----------
    start_days : int
        Date de début en jours depuis la date de référence
    end_days : int
        Date de fin en jours depuis la date de référence
    convention : str
        Convention de calcul (ACT/365, ACT/360, 30/360)
        
    Returns:
    --------
    float : Fraction d'année
    """
    if convention == 'ACT/365':
        return (end_days - start_days) / 365
    elif convention == 'ACT/360':
        return (end_days - start_days) / 360
    elif convention == '30/360':
        return (end_days - start_days) / 360
    else:
        raise ValueError(f"Convention non supportée: {convention}")

def prepare_data(df, rate_col='Rate'):
    """Préparation des données"""
    # Conversion des taux en format numérique
    df[rate_col] = df[rate_col].str.replace(',', '.').astype(float) / 100
    
    # Calcul des maturités en jours
    df['MaturityDays'] = df['Maturité'].apply(convert_maturity_to_days)
    
    # Calcul des dates de début
    df['StartDays'] = df['Date Debut'].apply(convert_maturity_to_days)
    
    # Calcul des dates d'échéance en jours à partir de la date de référence
    df['EndDays'] = df['StartDays'] + df['MaturityDays']
    
    return df

# Préparation des données
eonia = prepare_data(eonia.copy())
euribor3m = prepare_data(euribor3m.copy())
euribor6m = prepare_data(euribor6m.copy(), rate_col='Rates')

# Fonctions d'interpolation
def interpolate_zc(t, curve):
    """
    Interpolation linéaire entre les maturités pour les zéro-coupons.
    Extrapolation constante pour les valeurs hors de la courbe.
    
    Parameters:
    -----------
    t : float
        Temps en jours pour lequel on souhaite interpoler
    curve : array
        Tableau numpy contenant les couples (temps, taux ZC)
        
    Returns:
    --------
    float : Taux zéro-coupon interpolé
    """
    times = curve[:, 0]
    zc_values = curve[:, 1]
    
    if t <= times[0]:
        return zc_values[0]  # Extrapolation constante à gauche
    if t >= times[-1]:
        return zc_values[-1]  # Extrapolation constante à droite
    
    # Interpolation linéaire
    for i in range(len(times)-1):
        if times[i] <= t <= times[i+1]:
            alpha = (t - times[i]) / (times[i+1] - times[i])
            return zc_values[i] + alpha * (zc_values[i+1] - zc_values[i])
    
    return None  # Ne devrait jamais arriver

def get_discount_factor(t, zc_curve):
    """Calcule le facteur d'actualisation à partir de la courbe ZC"""
    zc_rate = interpolate_zc(t, zc_curve)
    return np.exp(-zc_rate * fractionAnnee(0, t, 'ACT/365'))

# Fonctions de valorisation des instruments
def depositRate(rate, start_days, end_days, convention='ACT/360'):
    """
    Calcule le facteur d'actualisation pour un dépôt.
    
    Parameters:
    -----------
    rate : float
        Taux du dépôt
    start_days : int
        Date de début en jours depuis la date de référence
    end_days : int
        Date de fin en jours depuis la date de référence
    convention : str
        Convention de calcul des jours
        
    Returns:
    --------
    float : Facteur d'actualisation
    """
    delta_t = fractionAnnee(start_days, end_days, convention)
    return 1 / (1 + rate * delta_t)

def fraRate(forward_rate, start_fra, end_fra, discount_curve=None, convention='ACT/360'):
    """
    Calcule les informations liées à un FRA.
    
    Parameters:
    -----------
    forward_rate : float
        Taux forward du FRA
    start_fra : int
        Date de début du FRA en jours
    end_fra : int
        Date de fin du FRA en jours
    discount_curve : array, optional
        Courbe d'actualisation pour l'approche multi-courbe
    convention : str
        Convention de calcul des jours
        
    Returns:
    --------
    float : Taux zéro-coupon implicite si mono-courbe,
            ou valeur implicite du FRA si multi-courbe
    """
    delta_t = fractionAnnee(start_fra, end_fra, convention)
    
    if discount_curve is None:
        # Approche mono-courbe
        df = 1 / (1 + forward_rate * delta_t)
        return df
    else:
        # Approche multi-courbe
        df_start = get_discount_factor(start_fra, discount_curve)
        df_end = get_discount_factor(end_fra, discount_curve)
        
        # Valeur du FRA
        fra_value = df_end * delta_t * (forward_rate - ((df_start / df_end - 1) / delta_t))
        return fra_value

def generate_swap_schedule(start_days, maturity_days, payment_frequency='1Y'):
    """
    Génère les dates de paiement pour un swap.
    
    Parameters:
    -----------
    start_days : int
        Date de début du swap en jours
    maturity_days : int
        Maturité du swap en jours
    payment_frequency : str
        Fréquence de paiement ('1Y', '6M', '3M', etc.)
        
    Returns:
    --------
    list : Liste des dates de paiement en jours
    """
    freq_days = convert_maturity_to_days(payment_frequency)
    dates = [start_days]
    
    current_day = start_days
    while current_day < start_days + maturity_days:
        current_day += freq_days
        if current_day <= start_days + maturity_days:
            dates.append(current_day)
    
    return dates

def swapRate(fixed_rate, start_days, maturity_days, fixed_freq='1Y', floating_freq='3M', 
            fixed_convention='ACT/360', floating_convention='ACT/360', 
            discount_curve=None, forward_curve=None):
    """
    Calcule la valeur d'un swap.
    
    Parameters:
    -----------
    fixed_rate : float
        Taux fixe du swap
    start_days : int
        Date de début du swap en jours
    maturity_days : int
        Maturité du swap en jours
    fixed_freq : str
        Fréquence de paiement de la jambe fixe
    floating_freq : str
        Fréquence de paiement de la jambe flottante
    fixed_convention, floating_convention : str
        Conventions de calcul des jours
    discount_curve : array, optional
        Courbe d'actualisation pour l'approche multi-courbe
    forward_curve : array, optional
        Courbe forward pour l'approche multi-courbe
        
    Returns:
    --------
    float : Valeur du swap
    """
    # Génération des dates de paiement
    fixed_dates = generate_swap_schedule(start_days, maturity_days, fixed_freq)
    floating_dates = generate_swap_schedule(start_days, maturity_days, floating_freq)
    
    if discount_curve is None and forward_curve is None:
        # Approche mono-courbe
        fixed_leg_value = 0
        for i in range(1, len(fixed_dates)):
            delta_t = fractionAnnee(fixed_dates[i-1], fixed_dates[i], fixed_convention)
            df = 1 / (1 + fixed_rate * fractionAnnee(0, fixed_dates[i], fixed_convention))
            fixed_leg_value += df * delta_t * fixed_rate
        
        # Pour un swap à l'initiation, la jambe variable vaut 1 - df_maturity
        df_maturity = 1 / (1 + fixed_rate * fractionAnnee(0, maturity_days, floating_convention))
        floating_leg_value = 1 - df_maturity
        
        return fixed_leg_value - floating_leg_value
    else:
        # Approche multi-courbe
        fixed_leg_value = 0
        for i in range(1, len(fixed_dates)):
            delta_t = fractionAnnee(fixed_dates[i-1], fixed_dates[i], fixed_convention)
            df = get_discount_factor(fixed_dates[i], discount_curve)
            fixed_leg_value += df * delta_t * fixed_rate
        
        floating_leg_value = 0
        for i in range(1, len(floating_dates)):
            delta_t = fractionAnnee(floating_dates[i-1], floating_dates[i], floating_convention)
            
            # Taux forward implicite
            df_start = get_discount_factor(floating_dates[i-1], forward_curve)
            df_end = get_discount_factor(floating_dates[i], forward_curve)
            forward_rate = (df_start / df_end - 1) / delta_t
            
            # Facteur d'actualisation
            df = get_discount_factor(floating_dates[i], discount_curve)
            
            floating_leg_value += df * delta_t * forward_rate
        
        return fixed_leg_value - floating_leg_value

# Fonction pour le stripping mono-courbe
def mono_curve_stripping(instruments, rate_col='Rate', conventions=None):
    """
    Réalise le stripping d'une courbe mono-courbe.
    
    Parameters:
    -----------
    instruments : DataFrame
        DataFrame contenant les instruments de marché
    rate_col : str
        Nom de la colonne contenant les taux
    conventions : dict
        Dictionnaire des conventions pour les différents instruments
        
    Returns:
    --------
    array : Tableau numpy contenant les couples (temps, taux ZC)
    """
    if conventions is None:
        conventions = {
            'DEPOSIT': 'ACT/360',
            'FRA': 'ACT/360',
            'SWAP': {
                'fixe': 'ACT/360',
                'variable': 'ACT/360'
            }
        }
    
    zc_curve = []
    
    # Tri des instruments par maturité
    instruments = instruments.sort_values(by='EndDays')
    
    for idx, instrument in instruments.iterrows():
        type_instr = instrument.get('Instrument', 'DEPOSIT')
        rate = instrument[rate_col]
        start_days = instrument['StartDays']
        end_days = instrument['EndDays']
        
        if type_instr == 'DEPOSIT':
            df = depositRate(rate, start_days, end_days, conventions['DEPOSIT'])
            # Conversion en taux ZC continu
            zc_rate = -np.log(df) / fractionAnnee(0, end_days, 'ACT/365')
            zc_curve.append((end_days, zc_rate))
            
        elif type_instr == 'FRA':
            # Si la courbe est vide, on ajoute d'abord un point initial
            if not zc_curve:
                raise ValueError("La courbe ZC doit contenir au moins un point avant d'ajouter un FRA")
            
            curve_array = np.array(zc_curve)
            
            def objective_func(zc_rate):
                # Calcul du facteur d'actualisation à la date d'échéance du FRA
                df_end = np.exp(-zc_rate * fractionAnnee(0, end_days, 'ACT/365'))
                
                # Calcul du taux forward implicite
                df_start = np.exp(-interpolate_zc(start_days, curve_array) * 
                                fractionAnnee(0, start_days, 'ACT/365'))
                
                forward_rate = (df_start / df_end - 1) / fractionAnnee(start_days, end_days, conventions['FRA'])
                
                # L'objectif est que le taux forward implicite soit égal au taux FRA
                return forward_rate - rate
            
            # On trouve le taux zéro-coupon qui donne le bon taux forward
            try:
                # Estimation initiale
                initial_guess = interpolate_zc(end_days, curve_array) 
                
                # Recherche de la racine
                zc_rate = newton(objective_func, initial_guess, tol=1e-10, maxiter=1000)
                zc_curve.append((end_days, zc_rate))
                
            except (ValueError, RuntimeError) as e:
                print(f"Erreur lors du calcul du FRA à {end_days} jours: {e}")
                # Fallback à une méthode plus robuste mais plus lente
                try:
                    zc_rate = brentq(objective_func, -0.1, 0.2, xtol=1e-10)
                    zc_curve.append((end_days, zc_rate))
                except ValueError as e:
                    print(f"Impossible de trouver une solution pour le FRA à {end_days} jours: {e}")
                
        elif type_instr == 'SWAP':
            curve_array = np.array(zc_curve)
            
            # Pour les swaps, on doit déterminer le taux zéro-coupon à la maturité
            # qui fait que la valeur du swap est nulle
            
            def objective_func(zc_rate):
                # On ajoute temporairement le nouveau point à la courbe
                temp_curve = np.vstack([curve_array, [end_days, zc_rate]])
                
                # Calcul de la valeur du swap
                fixed_freq = '1Y'  # À adapter selon le type de swap
                floating_freq = '1Y'  # À adapter selon le type de swap
                
                if 'EONIA' in str(instruments):
                    floating_freq = '1D'
                elif 'EURIBOR3M' in str(instruments):
                    floating_freq = '3M'
                elif 'EURIBOR6M' in str(instruments):
                    floating_freq = '6M'
                
                fixed_convention = conventions['SWAP']['fixe']
                floating_convention = conventions['SWAP']['variable']
                
                # Calcul via bootstrap du taux qui rendrait le swap à la par
                # On part du principe que le swap est à la par
                fixed_leg_value = 0
                for t in range(int(start_days), int(end_days)+1, 365):  # Approximation pour les paiements annuels
                    if t > start_days:  # Eviter le premier point qui est à 0
                        df = np.exp(-interpolate_zc(t, temp_curve) * fractionAnnee(0, t, 'ACT/365'))
                        fixed_leg_value += df
                
                # Valeur actuelle du notionnel à la maturité
                df_maturity = np.exp(-zc_rate * fractionAnnee(0, end_days, 'ACT/365'))
                floating_leg_value = 1 - df_maturity
                
                # Le taux swap est tel que les deux jambes sont équivalentes
                actual_swap_rate = rate  # Taux du swap observé sur le marché
                
                # Pour un swap à la par, la valeur actuelle des flux fixes = 1 - df_maturity
                implied_fixed_leg = fixed_leg_value * actual_swap_rate
                
                return implied_fixed_leg - floating_leg_value
            
            try:
                # Estimation initiale
                if len(curve_array) > 0:
                    initial_guess = interpolate_zc(end_days, curve_array)
                else:
                    initial_guess = 0.01  # Valeur arbitraire si la courbe est vide
                
                # Recherche de la racine
                zc_rate = newton(objective_func, initial_guess, tol=1e-10, maxiter=1000)
                zc_curve.append((end_days, zc_rate))
                
            except (ValueError, RuntimeError) as e:
                print(f"Erreur lors du calcul du SWAP à {end_days} jours: {e}")
                # Fallback à une méthode plus robuste mais plus lente
                try:
                    zc_rate = brentq(objective_func, -0.05, 0.15, xtol=1e-10)
                    zc_curve.append((end_days, zc_rate))
                except ValueError as e:
                    print(f"Impossible de trouver une solution pour le SWAP à {end_days} jours: {e}")
    
    # Tri par maturité
    zc_curve.sort(key=lambda x: x[0])
    
    return np.array(zc_curve)

# Fonction pour le stripping multi-courbe
def multi_curve_stripping(forward_instruments, discount_curve, rate_col='Rate', conventions=None):
    """
    Réalise le stripping d'une courbe forward en utilisant une courbe d'actualisation.
    
    Parameters:
    -----------
    forward_instruments : DataFrame
        DataFrame contenant les instruments de marché
    discount_curve : array
        Courbe d'actualisation
    rate_col : str
        Nom de la colonne contenant les taux
    conventions : dict
        Dictionnaire des conventions pour les différents instruments
        
    Returns:
    --------
    array : Tableau numpy contenant les couples (temps, taux ZC forward)
    """
    if conventions is None:
        conventions = {
            'DEPOSIT': 'ACT/360',
            'FRA': 'ACT/360',
            'SWAP': {
                'fixe': 'ACT/360',
                'variable': 'ACT/360'
            }
        }
    
    zc_forward_curve = []
    
    # Tri des instruments par maturité
    forward_instruments = forward_instruments.sort_values(by='EndDays')
    
    for idx, instrument in forward_instruments.iterrows():
        type_instr = instrument.get('Instrument', 'DEPOSIT')
        rate = instrument[rate_col]
        start_days = instrument['StartDays']
        end_days = instrument['EndDays']
        
        if type_instr == 'DEPOSIT':
            # Pour un dépôt, nous avons besoin du facteur d'actualisation de la courbe d'actualisation
            discount_df = get_discount_factor(end_days, discount_curve)
            
            # Calcul du taux forward implicite
            delta_t = fractionAnnee(start_days, end_days, conventions['DEPOSIT'])
            
            # Le taux du dépôt est directement le taux forward
            forward_df = discount_df / (1 + rate * delta_t)
            forward_zc = -np.log(forward_df) / fractionAnnee(0, end_days, 'ACT/365')
            
            zc_forward_curve.append((end_days, forward_zc))
            
        elif type_instr == 'FRA':
            # Si la courbe est vide, on ne peut pas procéder
            if not zc_forward_curve:
                raise ValueError("La courbe forward doit contenir au moins un point avant d'ajouter un FRA")
            
            forward_curve_array = np.array(zc_forward_curve)
            
            def objective_func(forward_zc_rate):
                # Courbe forward temporaire avec le nouveau point
                temp_forward_curve = np.vstack([forward_curve_array, [end_days, forward_zc_rate]])
                
                # Facteurs d'actualisation pour la courbe de discount
                discount_df_start = get_discount_factor(start_days, discount_curve)
                discount_df_end = get_discount_factor(end_days, discount_curve)
                
                # Taux forward implicite
                forward_df_start = np.exp(-interpolate_zc(start_days, temp_forward_curve) * 
                                        fractionAnnee(0, start_days, 'ACT/365'))
                forward_df_end = np.exp(-forward_zc_rate * fractionAnnee(0, end_days, 'ACT/365'))
                
                delta_t = fractionAnnee(start_days, end_days, conventions['FRA'])
                implied_rate = (forward_df_start / forward_df_end - 1) / delta_t
                
                # L'objectif est que le taux forward implicite soit égal au taux FRA
                return implied_rate - rate
            
            try:
                # Estimation initiale
                initial_guess = interpolate_zc(end_days, forward_curve_array)
                
                # Recherche de la racine
                forward_zc_rate = newton(objective_func, initial_guess, tol=1e-10, maxiter=1000)
                zc_forward_curve.append((end_days, forward_zc_rate))
                
            except (ValueError, RuntimeError) as e:
                print(f"Erreur lors du calcul du FRA à {end_days} jours: {e}")
                # Fallback à une méthode plus robuste mais plus lente
                try:
                    forward_zc_rate = brentq(objective_func, -0.1, 0.2, xtol=1e-10)
                    zc_forward_curve.append((end_days, forward_zc_rate))
                except ValueError as e:
                    print(f"Impossible de trouver une solution pour le FRA à {end_days} jours: {e}")
                    
        elif type_instr == 'SWAP':
            forward_curve_array = np.array(zc_forward_curve)
            
            def objective_func(forward_zc_rate):
                # Courbe forward temporaire avec le nouveau point
                temp_forward_curve = np.vstack([forward_curve_array, [end_days, forward_zc_rate]])
                
                # Paramètres du swap
                fixed_freq = '1Y'  # À adapter selon le type de swap
                
                if 'EURIBOR3M' in str(forward_instruments):
                    floating_freq = '3M'
                elif 'EURIBOR6M' in str(forward_instruments):
                    floating_freq = '6M'
                else:
                    floating_freq = '1Y'  # Valeur par défaut
                
                fixed_convention = conventions['SWAP']['fixe']
                floating_convention = conventions['SWAP']['variable']
                
                # Calcul des flux fixes
                fixed_leg_value = 0
                for t in range(int(start_days), int(end_days)+1, 365):  # Approximation pour les paiements annuels
                    if t > start_days:  # Eviter le premier point qui est à 0
                        df = get_discount_factor(t, discount_curve)
                        delta_t = 1  # Approximation pour un an
                        fixed_leg_value += df * delta_t * rate
                
                # Calcul des flux variables
                floating_leg_value = 0
                for i in range(int(start_days), int(end_days), 90 if floating_freq == '3M' else 180):  # Approximation pour 3M ou 6M
                    next_i = min(i + (90 if floating_freq == '3M' else 180), int(end_days))
                    if i > start_days and next_i <= end_days:  # Eviter le premier point et ne pas dépasser la maturité
                        # Taux forward implicite
                        forward_df_start = np.exp(-interpolate_zc(i, temp_forward_curve) * 
                                                fractionAnnee(0, i, 'ACT/365'))
                        forward_df_end = np.exp(-interpolate_zc(next_i, temp_forward_curve) * 
                                                fractionAnnee(0, next_i, 'ACT/365'))
                        
                        delta_t = fractionAnnee(i, next_i, floating_convention)
                        forward_rate = (forward_df_start / forward_df_end - 1) / delta_t
                        
                        # Facteur d'actualisation
                        df = get_discount_factor(next_i, discount_curve)
                        
                        floating_leg_value += df * delta_t * forward_rate
                
                # Pour un swap à la par, les deux jambes doivent avoir la même valeur
                return fixed_leg_value - floating_leg_value
            
            try:
                # Estimation initiale
                initial_guess = interpolate_zc(end_days, forward_curve_array) if len(forward_curve_array) > 0 else 0.01
                
                # Recherche de la racine
                forward_zc_rate = newton(objective_func, initial_guess, tol=1e-10, maxiter=1000)
                zc_forward_curve.append((end_days, forward_zc_rate))
                
            except (ValueError, RuntimeError) as e:
                print(f"Erreur lors du calcul du SWAP à {end_days} jours: {e}")
                # Fallback à une méthode plus robuste mais plus lente
                try:
                    forward_zc_rate = brentq(objective_func, -0.05, 0.15, xtol=1e-10)
                    zc_forward_curve.append((end_days, forward_zc_rate))
                except ValueError as e:
                    print(f"Impossible de trouver une solution pour le SWAP à {end_days} jours: {e}")
    
    # Tri par maturité
    zc_forward_curve.sort(key=lambda x: x[0])
    
    return np.array(zc_forward_curve)

# Fonction pour vérifier la valorisation des instruments

def verify_curve(instruments, zc_curve, discount_curve=None, rate_col='Rate', conventions=None):
    """
    Vérifie que les instruments de marché sont bien valorisés avec la courbe ZC.
    
    Parameters:
    -----------
    instruments : DataFrame
        DataFrame contenant les instruments de marché
    zc_curve : array
        Courbe ZC à vérifier
    discount_curve : array, optional
        Courbe d'actualisation pour l'approche multi-courbe
    rate_col : str
        Nom de la colonne contenant les taux
    conventions : dict
        Dictionnaire des conventions pour les différents instruments
        
    Returns:
    --------
    DataFrame : Tableau des erreurs
    """
    if conventions is None:
        conventions = {
            'DEPOSIT': 'ACT/360',
            'FRA': 'ACT/360',
            'SWAP': {
                'fixe': 'ACT/360',
                'variable': 'ACT/360'
            }
        }
    
    errors = []
    
    for idx, instrument in instruments.iterrows():
        type_instr = instrument.get('Instrument', 'DEPOSIT')
        rate = instrument[rate_col]
        start_days = instrument['StartDays']
        end_days = instrument['EndDays']
        
        if type_instr == 'DEPOSIT':
            if discount_curve is None:
                # Mono-courbe
                zc_rate = interpolate_zc(end_days, zc_curve)
                df = np.exp(-zc_rate * fractionAnnee(0, end_days, 'ACT/365'))
                delta_t = fractionAnnee(start_days, end_days, conventions['DEPOSIT'])
                implied_rate = (1/df - 1) / delta_t
                
            else:
                # Multi-courbe
                discount_df = get_discount_factor(end_days, discount_curve)
                forward_zc = interpolate_zc(end_days, zc_curve)
                forward_df = np.exp(-forward_zc * fractionAnnee(0, end_days, 'ACT/365'))
                
                delta_t = fractionAnnee(start_days, end_days, conventions['DEPOSIT'])
                implied_rate = (1/discount_df - 1) / delta_t
            
            error = rate - implied_rate
            errors.append((type_instr, end_days, rate, implied_rate, error))
            
        elif type_instr == 'FRA':
            if discount_curve is None:
                # Mono-courbe
                start_zc = interpolate_zc(start_days, zc_curve)
                end_zc = interpolate_zc(end_days, zc_curve)
                
                df_start = np.exp(-start_zc * fractionAnnee(0, start_days, 'ACT/365'))
                df_end = np.exp(-end_zc * fractionAnnee(0, end_days, 'ACT/365'))
                
                delta_t = fractionAnnee(start_days, end_days, conventions['FRA'])
                implied_rate = (df_start / df_end - 1) / delta_t
                
            else:
                # Multi-courbe
                # Facteurs d'actualisation de la courbe discount
                discount_df_start = get_discount_factor(start_days, discount_curve)
                discount_df_end = get_discount_factor(end_days, discount_curve)
                
                # Facteurs d'actualisation de la courbe forward
                forward_start_zc = interpolate_zc(start_days, zc_curve)
                forward_end_zc = interpolate_zc(end_days, zc_curve)
                
                forward_df_start = np.exp(-forward_start_zc * fractionAnnee(0, start_days, 'ACT/365'))
                forward_df_end = np.exp(-forward_end_zc * fractionAnnee(0, end_days, 'ACT/365'))
                
                delta_t = fractionAnnee(start_days, end_days, conventions['FRA'])
                implied_rate = (forward_df_start / forward_df_end - 1) / delta_t
            
            error = rate - implied_rate
            errors.append((type_instr, end_days, rate, implied_rate, error))
            
        elif type_instr == 'SWAP':
            if discount_curve is None:
                # Mono-courbe : on calcule la valeur du swap avec le taux observé
                # et on vérifie que la valeur est proche de zéro
                
                # Approximation simple : on calcule le taux zéro-coupon à la maturité
                # et on en déduit le taux swap implicite
                
                end_zc = interpolate_zc(end_days, zc_curve)
                df_end = np.exp(-end_zc * fractionAnnee(0, end_days, 'ACT/365'))
                
                # Estimation très simplifiée du taux swap
                duration = 0
                annuity = 0
                
                # Calcul d'une annuité simplifiée
                for t in range(int(start_days), int(end_days)+1, 365):  # Paiements annuels approximatifs
                    if t > start_days:
                        t_zc = interpolate_zc(t, zc_curve)
                        df = np.exp(-t_zc * fractionAnnee(0, t, 'ACT/365'))
                        duration += t * df
                        annuity += df
                
                if annuity > 0:
                    duration /= annuity
                    implied_rate = (1 - df_end) / annuity
                else:
                    implied_rate = rate  # Fallback si annuité nulle
                
            else:
                # Multi-courbe
                fixed_freq = '1Y'
                
                if 'EURIBOR3M' in str(instruments):
                    floating_freq = '3M'
                elif 'EURIBOR6M' in str(instruments):
                    floating_freq = '6M'
                else:
                    floating_freq = '1Y'
                
                fixed_convention = conventions['SWAP']['fixe']
                floating_convention = conventions['SWAP']['variable']
                
                # Calcul des flux fixes
                fixed_leg_value = 0
                for t in range(int(start_days), int(end_days)+1, 365):
                    if t > start_days:
                        df = get_discount_factor(t, discount_curve)
                        delta_t = 1  # Approximation pour un an
                        fixed_leg_value += df * delta_t * rate
                
                # Calcul des flux variables
                floating_leg_value = 0
                freq_days = 90 if floating_freq == '3M' else 180  # 3M ou 6M
                
                for i in range(int(start_days), int(end_days), freq_days):
                    next_i = min(i + freq_days, int(end_days))
                    if i > start_days and next_i <= end_days:
                        # Taux forward implicite
                        forward_start_zc = interpolate_zc(i, zc_curve)
                        forward_end_zc = interpolate_zc(next_i, zc_curve)
                        
                        forward_df_start = np.exp(-forward_start_zc * fractionAnnee(0, i, 'ACT/365'))
                        forward_df_end = np.exp(-forward_end_zc * fractionAnnee(0, next_i, 'ACT/365'))
                        
                        delta_t = fractionAnnee(i, next_i, floating_convention)
                        forward_rate = (forward_df_start / forward_df_end - 1) / delta_t
                        
                        # Facteur d'actualisation
                        df = get_discount_factor(next_i, discount_curve)
                        
                        floating_leg_value += df * delta_t * forward_rate
                
                # Le taux swap implicite est tel que les deux jambes sont équivalentes
                implied_rate = floating_leg_value / fixed_leg_value
            
            error = rate - implied_rate
            errors.append((type_instr, end_days, rate, implied_rate, error))
    
    # Conversion en DataFrame
    return pd.DataFrame(errors, columns=['Instrument', 'Maturité (jours)', 'Taux marché', 'Taux implicite', 'Erreur'])

# Fonction pour visualiser les courbes
def plot_curves(curves_data, title='Courbes de taux zéro-coupon', ylabel='Taux (%)', save_path=None):
    """
    Visualise les courbes zéro-coupon.
    
    Parameters:
    -----------
    curves_data : dict
        Dictionnaire contenant les courbes à visualiser : {'nom_courbe': array}
    title : str
        Titre du graphique
    ylabel : str
        Étiquette de l'axe y
    save_path : str, optional
        Chemin pour sauvegarder l'image
    """
    plt.figure(figsize=(12, 8))
    
    for curve_name, curve_array in curves_data.items():
        # Conversion des jours en années
        days = curve_array[:, 0]
        years = days / 365
        
        # Affichage des taux en pourcentage
        rates = curve_array[:, 1] * 100
        
        plt.plot(years, rates, marker='o', linestyle='-', label=curve_name)
    
    plt.title(title)
    plt.xlabel('Années')
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.legend()
    
    if save_path:
        plt.savefig(save_path)
    
    plt.show()

# Définir les conventions spécifiques pour chaque courbe selon l'énoncé
conventions_eonia = {
    'DEPOSIT': 'ACT/360',
    'FRA': 'ACT/360',
    'SWAP': {
        'fixe': 'ACT/360',
        'variable': 'ACT/360'
    }
}

conventions_euribor3m = {
    'DEPOSIT': 'ACT/360',
    'FRA': 'ACT/360',
    'SWAP': {
        'fixe': 'ACT/360',
        'variable': '30/360'
    }
}

conventions_euribor6m = {
    'DEPOSIT': 'ACT/360',
    'FRA': 'ACT/360',
    'SWAP': {
        'fixe': 'ACT/360',
        'variable': '30/360'
    }
}

# Exécution du stripping mono-courbe pour les 3 courbes
print("Stripping mono-courbe en cours...")

# EONIA
zc_eonia = mono_curve_stripping(eonia, rate_col='Rate', conventions=conventions_eonia)
print(f"Courbe EONIA : {len(zc_eonia)} points")

# EURIBOR 3M
zc_euribor3m = mono_curve_stripping(euribor3m, rate_col='Rate', conventions=conventions_euribor3m)
print(f"Courbe EURIBOR 3M : {len(zc_euribor3m)} points")

# EURIBOR 6M
zc_euribor6m = mono_curve_stripping(euribor6m, rate_col='Rates', conventions=conventions_euribor6m)
print(f"Courbe EURIBOR 6M : {len(zc_euribor6m)} points")

# Visualisation des courbes mono-courbe
mono_curves = {
    'EONIA': zc_eonia,
    'EURIBOR 3M': zc_euribor3m,
    'EURIBOR 6M': zc_euribor6m
}

plot_curves(mono_curves, title='Courbes de taux zéro-coupon (Mono-courbe)', ylabel='Taux ZC (%)')

# Vérification de la valorisation des instruments avec les courbes mono-courbe
print("\nVérification des valorisations (Mono-courbe):")
print("\nEONIA:")
print(verify_curve(eonia, zc_eonia, rate_col='Rate', conventions=conventions_eonia))
print("\nEURIBOR 3M:")
print(verify_curve(euribor3m, zc_euribor3m, rate_col='Rate', conventions=conventions_euribor3m))
print("\nEURIBOR 6M:")
print(verify_curve(euribor6m, zc_euribor6m, rate_col='Rates', conventions=conventions_euribor6m))

# Exécution du stripping multi-courbe avec EONIA comme courbe d'actualisation
print("\nStripping multi-courbe en cours...")

# EURIBOR 3M avec EONIA comme courbe d'actualisation
zc_euribor3m_multi = multi_curve_stripping(euribor3m, zc_eonia, rate_col='Rate', conventions=conventions_euribor3m)
print(f"Courbe EURIBOR 3M (multi) : {len(zc_euribor3m_multi)} points")

# EURIBOR 6M avec EONIA comme courbe d'actualisation
zc_euribor6m_multi = multi_curve_stripping(euribor6m, zc_eonia, rate_col='Rates', conventions=conventions_euribor6m)
print(f"Courbe EURIBOR 6M (multi) : {len(zc_euribor6m_multi)} points")

# Visualisation des courbes multi-courbe
multi_curves = {
    'EONIA (discount)': zc_eonia,
    'EURIBOR 3M (forward)': zc_euribor3m_multi,
    'EURIBOR 6M (forward)': zc_euribor6m_multi
}

plot_curves(multi_curves, title='Courbes de taux zéro-coupon (Multi-courbe avec EONIA comme discount)', ylabel='Taux ZC (%)')

# Visualisation des spreads entre les courbes mono et multi
spreads_3m = np.array([[zc_euribor3m_multi[i, 0], zc_euribor3m_multi[i, 1] - interpolate_zc(zc_euribor3m_multi[i, 0], zc_eonia)]
                      for i in range(len(zc_euribor3m_multi))])

spreads_6m = np.array([[zc_euribor6m_multi[i, 0], zc_euribor6m_multi[i, 1] - interpolate_zc(zc_euribor6m_multi[i, 0], zc_eonia)]
                      for i in range(len(zc_euribor6m_multi))])

spreads = {
    'EURIBOR 3M - EONIA': spreads_3m,
    'EURIBOR 6M - EONIA': spreads_6m
}

plot_curves(spreads, title='Spreads des courbes forward par rapport à EONIA', ylabel='Spread ZC (%)')

# Vérification de la valorisation des instruments avec les courbes multi-courbe
print("\nVérification des valorisations (Multi-courbe):")
print("\nEURIBOR 3M (avec EONIA comme discount):")
print(verify_curve(euribor3m, zc_euribor3m_multi, discount_curve=zc_eonia, rate_col='Rate', conventions=conventions_euribor3m))
print("\nEURIBOR 6M (avec EONIA comme discount):")
print(verify_curve(euribor6m, zc_euribor6m_multi, discount_curve=zc_eonia, rate_col='Rates', conventions=conventions_euribor6m))

# Comparaison des taux forward implicites
def calculate_forward_rates(zc_curve, convention='ACT/360'):
    """Calcule les taux forward implicites à partir de la courbe ZC"""
    forwards = []
    
    # Trier par maturité croissante
    zc_curve = zc_curve[zc_curve[:, 0].argsort()]
    
    for i in range(1, len(zc_curve)):
        start_days = zc_curve[i-1, 0]
        end_days = zc_curve[i, 0]
        
        start_zc = zc_curve[i-1, 1]
        end_zc = zc_curve[i, 1]
        
        df_start = np.exp(-start_zc * fractionAnnee(0, start_days, 'ACT/365'))
        df_end = np.exp(-end_zc * fractionAnnee(0, end_days, 'ACT/365'))
        
        delta_t = fractionAnnee(start_days, end_days, convention)
        forward_rate = (df_start / df_end - 1) / delta_t
        
        forwards.append((end_days, forward_rate, start_days))
    
    return forwards

# Calcul des taux forward implicites
forwards_eonia = calculate_forward_rates(zc_eonia)
forwards_euribor3m = calculate_forward_rates(zc_euribor3m)
forwards_euribor6m = calculate_forward_rates(zc_euribor6m)
forwards_euribor3m_multi = calculate_forward_rates(zc_euribor3m_multi)
forwards_euribor6m_multi = calculate_forward_rates(zc_euribor6m_multi)

# Synthèse des résultats
print("\nSynthèse des courbes zéro-coupon:")
for curve_name, curve in [('EONIA', zc_eonia), ('EURIBOR 3M', zc_euribor3m), ('EURIBOR 6M', zc_euribor6m),
                        ('EURIBOR 3M (multi)', zc_euribor3m_multi), ('EURIBOR 6M (multi)', zc_euribor6m_multi)]:
    days = curve[:, 0]
    rates = curve[:, 1] * 100  # Conversion en pourcentage
    print(f"\n{curve_name}:")
    for i, (day, rate) in enumerate(zip(days, rates)):
        year = day / 365
        print(f"  {year:.2f} an(s): {rate:.4f}%")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# On suppose que les courbes zc_euribor3m, zc_euribor3m_multi, zc_euribor6m, zc_euribor6m_multi 
# sont déjà calculées à partir du code fourni

# Fonction pour calculer les taux forward successifs toutes les semaines
def calculate_sequential_forwards(zc_curve, forward_period_days, step_days=7, max_days=365*30, convention='ACT/360'):
    """
    Calcule les taux forward successifs avec un pas régulier.
    
    Parameters:
    -----------
    zc_curve : array
        Courbe zéro-coupon (temps en jours, taux)
    forward_period_days : int
        Période du forward (ex: 90 jours pour un 3M, 180 jours pour un 6M)
    step_days : int
        Pas entre les fixings (ex: 7 jours pour chaque semaine)
    max_days : int
        Horizon maximal en jours
    convention : str
        Convention de calcul des jours
        
    Returns:
    --------
    list : Liste de tuples (date_fixing, taux_forward, date_début, date_fin)
    """
    forwards = []
    
    # Fonction pour interpoler les taux zéro-coupon
    def interpolate_zc(t, curve):
        times = curve[:, 0]
        zc_values = curve[:, 1]
        
        if t <= times[0]:
            return zc_values[0]  # Extrapolation constante à gauche
        if t >= times[-1]:
            return zc_values[-1]  # Extrapolation constante à droite
        
        # Interpolation linéaire
        for i in range(len(times)-1):
            if times[i] <= t <= times[i+1]:
                alpha = (t - times[i]) / (times[i+1] - times[i])
                return zc_values[i] + alpha * (zc_values[i+1] - zc_values[i])
        
        return None
    
    # Fonction pour calculer la fraction d'année
    def fractionAnnee(start_days, end_days, convention='ACT/365'):
        if convention == 'ACT/365':
            return (end_days - start_days) / 365
        elif convention == 'ACT/360':
            return (end_days - start_days) / 360
        elif convention == '30/360':
            return (end_days - start_days) / 360
        else:
            raise ValueError(f"Convention non supportée: {convention}")
    
    # Calcul des forwards pour chaque fixing
    for fixing_day in range(0, max_days, step_days):
        start_day = fixing_day
        end_day = fixing_day + forward_period_days
        
        # Si la date de fin dépasse notre horizon, on s'arrête
        if end_day > max_days:
            break
        
        # Interpolation des taux zéro-coupon pour les dates de début et de fin
        start_zc = interpolate_zc(start_day, zc_curve)
        end_zc = interpolate_zc(end_day, zc_curve)
        
        # Calcul des facteurs d'actualisation
        df_start = np.exp(-start_zc * fractionAnnee(0, start_day, 'ACT/365'))
        df_end = np.exp(-end_zc * fractionAnnee(0, end_day, 'ACT/365'))
        
        # Calcul du taux forward
        delta_t = fractionAnnee(start_day, end_day, convention)
        forward_rate = (df_start / df_end - 1) / delta_t
        
        forwards.append((fixing_day, forward_rate, start_day, end_day))
    
    return forwards

# Calcul des taux forward 3M successifs (fixing toutes les semaines jusqu'à 30 ans)
forward_3m_mono = calculate_sequential_forwards(zc_euribor3m, 90, 7, 365*30)
forward_3m_multi = calculate_sequential_forwards(zc_euribor3m_multi, 90, 7, 365*30)

# Calcul des taux forward 6M successifs (fixing toutes les semaines jusqu'à 30 ans)
forward_6m_mono = calculate_sequential_forwards(zc_euribor6m, 180, 7, 365*30)
forward_6m_multi = calculate_sequential_forwards(zc_euribor6m_multi, 180, 7, 365*30)

# Conversion en DataFrame pour visualisation
df_3m_mono = pd.DataFrame(forward_3m_mono, columns=['FixingDay', 'ForwardRate', 'StartDay', 'EndDay'])
df_3m_multi = pd.DataFrame(forward_3m_multi, columns=['FixingDay', 'ForwardRate', 'StartDay', 'EndDay'])
df_6m_mono = pd.DataFrame(forward_6m_mono, columns=['FixingDay', 'ForwardRate', 'StartDay', 'EndDay'])
df_6m_multi = pd.DataFrame(forward_6m_multi, columns=['FixingDay', 'ForwardRate', 'StartDay', 'EndDay'])

# Conversion des jours en années pour l'affichage
df_3m_mono['FixingYear'] = df_3m_mono['FixingDay'] / 365
df_3m_multi['FixingYear'] = df_3m_multi['FixingDay'] / 365
df_6m_mono['FixingYear'] = df_6m_mono['FixingDay'] / 365
df_6m_multi['FixingYear'] = df_6m_multi['FixingDay'] / 365

# Conversion des taux en pourcentage
df_3m_mono['ForwardRate%'] = df_3m_mono['ForwardRate'] * 100
df_3m_multi['ForwardRate%'] = df_3m_multi['ForwardRate'] * 100
df_6m_mono['ForwardRate%'] = df_6m_mono['ForwardRate'] * 100
df_6m_multi['ForwardRate%'] = df_6m_multi['ForwardRate'] * 100

# Visualisation des taux forward 3M
plt.figure(figsize=(14, 8))
plt.plot(df_3m_mono['FixingYear'], df_3m_mono['ForwardRate%'], 'b-', label='EURIBOR 3M Mono-courbe')
plt.plot(df_3m_multi['FixingYear'], df_3m_multi['ForwardRate%'], 'r-', label='EURIBOR 3M Multi-courbe')
plt.title('Taux Forward 3M EURIBOR (fixing hebdomadaire)')
plt.xlabel('Années')
plt.ylabel('Taux Forward 3M (%)')
plt.grid(True)
plt.legend()
plt.savefig('forward_3m_comparison.png')
plt.show()

# Visualisation des taux forward 6M
plt.figure(figsize=(14, 8))
plt.plot(df_6m_mono['FixingYear'], df_6m_mono['ForwardRate%'], 'g-', label='EURIBOR 6M Mono-courbe')
plt.plot(df_6m_multi['FixingYear'], df_6m_multi['ForwardRate%'], 'm-', label='EURIBOR 6M Multi-courbe')
plt.title('Taux Forward 6M EURIBOR (fixing hebdomadaire)')
plt.xlabel('Années')
plt.ylabel('Taux Forward 6M (%)')
plt.grid(True)
plt.legend()
plt.savefig('forward_6m_comparison.png')
plt.show()

# Calcul et visualisation des spreads entre mono-courbe et multi-courbe
spread_3m = df_3m_multi['ForwardRate%'] - df_3m_mono['ForwardRate%']
spread_6m = df_6m_multi['ForwardRate%'] - df_6m_mono['ForwardRate%']

plt.figure(figsize=(14, 8))
plt.plot(df_3m_mono['FixingYear'], spread_3m, 'b-', label='Spread 3M (Multi - Mono)')
plt.plot(df_6m_mono['FixingYear'], spread_6m, 'g-', label='Spread 6M (Multi - Mono)')
plt.title('Spreads entre approches Multi-courbe et Mono-courbe')
plt.xlabel('Années')
plt.ylabel('Spread (%)')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
plt.savefig('forward_spreads.png')
plt.show()

# Analyse statistique des spreads
print("\nAnalyse statistique du spread 3M (Multi - Mono):")
print(f"Moyenne: {spread_3m.mean():.6f}%")
print(f"Écart-type: {spread_3m.std():.6f}%")
print(f"Min: {spread_3m.min():.6f}%")
print(f"Max: {spread_3m.max():.6f}%")

print("\nAnalyse statistique du spread 6M (Multi - Mono):")
print(f"Moyenne: {spread_6m.mean():.6f}%")
print(f"Écart-type: {spread_6m.std():.6f}%")
print(f"Min: {spread_6m.min():.6f}%")
print(f"Max: {spread_6m.max():.6f}%")

# Analyse de la structure à terme des spreads
plt.figure(figsize=(14, 8))
plt.plot(df_3m_mono['FixingYear'], df_3m_mono['ForwardRate%'], 'b--', alpha=0.5, label='3M Mono')
plt.plot(df_3m_multi['FixingYear'], df_3m_multi['ForwardRate%'], 'b-', label='3M Multi')
plt.plot(df_6m_mono['FixingYear'], df_6m_mono['ForwardRate%'], 'g--', alpha=0.5, label='6M Mono')
plt.plot(df_6m_multi['FixingYear'], df_6m_multi['ForwardRate%'], 'g-', label='6M Multi')
plt.title('Comparaison des taux forward 3M et 6M (Mono vs Multi)')
plt.xlabel('Années')
plt.ylabel('Taux Forward (%)')
plt.grid(True)
plt.legend()
plt.savefig('all_forward_comparison.png')
plt.show()
```
